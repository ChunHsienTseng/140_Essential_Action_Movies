## Return A List Of URLs Of Each Movie

```python
list_url = [i.find('a').get('href') for i in name]

list_url
```

## Return Tomato Score, Audience Score, Movie Rating

```python
i = 0
list_tm_score, list_ad_score, list_rating = [], [], []

for url in list_url:
    response = requests.get(url = url)
    print(f"url #{i}: {url}, {response.status_code}")
    html = response.content
    soup = BeautifulSoup(html, 'lxml')
    score_board = soup.find_all(name = 'score-board-deprecated')
    tm_score = int(score_board[0].get('tomatometerscore'))
    ad_score = int(score_board[0].get('audiencescore'))
    rating = score_board[0].get('rating')
    list_tm_score.append(tm_score)
    list_ad_score.append(ad_score)
    list_rating.append(rating)
    i += 1

print(len(list_tm_score), len(list_ad_score), len(list_rating))
```

## Return Genre

```python
i = 0
list_genre = []

for url in list_url:
    response = requests.get(url = url)
    print(f"url #{i}: {url}, {response.status_code}")
    html = response.content
    soup = BeautifulSoup(html, 'lxml')
    span_genre = soup.find_all(name = 'span', attrs = {'class': 'genre'})
    genre = [i.text.replace(' ', '').replace('\n', '').replace(',', ', ') for i in span_genre]
    list_genre.extend(genre)
    i += 1

list_genre
```

### seperate each genre

```python
list_genre = [j for i in list_genre for j in i.split(', ')]

list_genre
```

## Return Original Language

```python
i = 0
list_lang = []

for url in list_url:
    response = requests.get(url = url)
    print(f"url #{i}: {url}, {response.status_code}")
    html = response.content
    soup = BeautifulSoup(html, 'lxml')
    span_li = soup.find_all(name = 'li', attrs = {'data-qa': 'movie-info-item'})
    lang = [i.text.strip() for i in span_li]
    lang = [i for i in lang if 'Original Language' in i][0][19:]
    list_lang.append(lang)
    i += 1

list_lang
```

## Return Director

```python
i = 0
list_director = []

for url in list_url:
    response = requests.get(url = url)
    print(f"url #{i}: {url}, {response.status_code}")
    html = response.content
    soup = BeautifulSoup(html, 'lxml')
    span_li = soup.find_all(name = 'li', attrs = {'data-qa': 'movie-info-item'})
    director = [i.text.strip() for i in span_li]
    director = [i[9:].replace('\n', '').replace('        ', '') for i in director if 'Director' in i]
    list_director.extend(director)
    i += 1

list_director
```

### seperate each director

```python
list_director = [j for i in list_director for j in i.split(', ')]

list_director
```

## Return Writer

```python
i = 0
list_writer = []

for url in list_url:
    response = requests.get(url = url)
    print(f"url #{i}: {url}, {response.status_code}")
    html = response.content
    soup = BeautifulSoup(html, 'lxml')
    span_li = soup.find_all(name = 'li', attrs = {'data-qa': 'movie-info-item'})
    writer = [i.text.strip() for i in span_li]
    writer = [i[7:].strip().replace('\n', '').replace('        ', '') for i in writer if 'Writer' in i]
    list_writer.extend(writer)
    i += 1

list_writer
```

### seperate each writer

```python
list_writer = [j for i in list_writer for j in i.split(', ')]

list_writer
```

## Return Synopsis

```python
i = 0
list_syn = []

for url in list_url:
    response = requests.get(url = url)
    print(f"url #{i}: {url}, {response.status_code}")
    html = response.content
    soup = BeautifulSoup(html, 'lxml')
    p_data = soup.find_all(name = 'p', attrs = {'data-qa': 'movie-info-synopsis'})
    syn = p_data[0].text.strip()
    list_syn.append(syn)
    i += 1

list_syn
```
